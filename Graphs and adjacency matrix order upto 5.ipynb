#### Spectral Analysis of All Non-Isomorphic Connected Simple Graphs up to Order 4

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import itertools

# --------- CHECK GRAPH ISOMORPHISM ---------
def is_new_graph(G, graph_list):
    """Check if G is isomorphic to any graph in graph_list"""
    for H in graph_list:
        if nx.is_isomorphic(G, H):
            return False
    return True

# --------- GENERATE NON-ISOMORPHIC CONNECTED GRAPHS ---------
def generate_nonisomorphic_connected_graphs(n):
    nodes = list(range(1, n+1))
    all_possible_edges = list(itertools.combinations(nodes, 2))
    
    unique_graphs = []
    
    for r in range(len(all_possible_edges) + 1):
        for edges in itertools.combinations(all_possible_edges, r):
            G = nx.Graph()
            G.add_nodes_from(nodes)
            G.add_edges_from(edges)
            
            if nx.is_connected(G):
                if is_new_graph(G, unique_graphs):
                    unique_graphs.append(G)
    
    return unique_graphs

# --------- MAIN ---------
for n in range(1, 5):  # up to order 4
    print(f"\n========== Order {n} ==========")
    
    graphs = generate_nonisomorphic_connected_graphs(n)
    print(f"Number of NON-ISOMORPHIC connected graphs: {len(graphs)}")
    
    for i, G in enumerate(graphs, start=1):
        print(f"\nGraph {i}")
        print("Edges:", list(G.edges()))
        
        # Adjacency Matrix
        A = nx.to_numpy_array(G, dtype=int)
        print("Adjacency Matrix:\n", A)
        
        # Eigenvalues
        eigenvalues = np.linalg.eigvals(A)
        print("Eigenvalues:\n", np.round(eigenvalues, 4))
        
        # Plot
        plt.figure(figsize=(3,3))
        pos = nx.spring_layout(G, seed=42)
        nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='black')
        plt.title(f"Order {n} - Graph {i}")
        plt.show()
```

### Spectral Analysis of All 21 Non-Isomorphic Connected Simple Graphs of Order 5

```python

# --------- GET ALL GRAPHS FROM ATLAS ---------
atlas = nx.graph_atlas_g()

# --------- FILTER: ORDER 5 + CONNECTED ---------
graphs_order_5 = [G for G in atlas if len(G.nodes()) == 5 and nx.is_connected(G)]

print("Total connected non-isomorphic graphs of order 5:", len(graphs_order_5))

# --------- PROCESS EACH GRAPH ---------
for i, G in enumerate(graphs_order_5, start=1):
    print(f"\n========== Graph {i} ==========")
    
    edges = list(G.edges())
    print("Edges:", edges)
    
    # Adjacency matrix
    A = nx.to_numpy_array(G, dtype=int)
    print("Adjacency Matrix:\n", A)
    
    # Eigenvalues
    eigenvalues = np.linalg.eigvals(A)
    print("Eigenvalues:\n", np.round(eigenvalues, 4))
    
    # Plot
    plt.figure(figsize=(3,3))
    pos = nx.spring_layout(G, seed=42)
    nx.draw(G, pos, with_labels=True, node_color='lightgreen', edge_color='black')
    plt.title(f"Graph {i}")
    plt.show()
```
